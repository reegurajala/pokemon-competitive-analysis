```python
import requests
import pandas as pd
import time
```

```python
pokemon_list = []

for i in range(1, 1026):
    response = requests.get(f"https://pokeapi.co/api/v2/pokemon-species/{i}")
    species_data = response.json()
    
    response2 = requests.get(f"https://pokeapi.co/api/v2/pokemon/{i}")
    data = response2.json()
    
    stats = {s['stat']['name']: s['base_stat'] for s in data['stats']}
    types = [t['type']['name'] for t in data['types']]
    
    pokemon_list.append({
        'id': data['id'],
        'name': data['name'],
        'type1': types[0],
        'type2': types[1] if len(types) > 1 else None,
        'hp': stats['hp'],
        'attack': stats['attack'],
        'defense': stats['defense'],
        'sp_attack': stats['special-attack'],
        'sp_defense': stats['special-defense'],
        'speed': stats['speed'],
        'is_legendary': species_data['is_legendary'],
        'is_mythical': species_data['is_mythical']
    })
    
    time.sleep(0.2)

print(f"Done! Scraped {len(pokemon_list)} Pokemon")
```

```python
df = pd.DataFrame(pokemon_list)
df.to_csv('pokemon_data.csv', index=False)
print(df[['name', 'is_legendary', 'is_mythical']].head(10))
```
